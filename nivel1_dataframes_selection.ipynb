```python
import pandas as pd
import numpy as np

df = pd.read_csv("financials_clean.csv")
df.head()
```

```python
print(df.isnull().sum())
```

```python
print(f"\nDuplicate rows: {df.duplicated().sum()}")
```

```python
print(df[df['Discount Band'].isnull()][['Segment', 'Product', 'Sales', 'Discount Band']].head(10))
```

```python
null_discount = df[df['Discount Band'].isnull()]
print(null_discount['Discounts'].describe())
print(f"\nDiscounts = 0: {(null_discount['Discounts'] == 0).sum()}")
print(f"Discounts > 0: {(null_discount['Discounts'] > 0).sum()}")
```

```python
df['Discount Band'] = df['Discount Band'].fillna('None')

# Confirmar
print(df['Discount Band'].value_counts())
print(f"\nNulls remaining: {df['Discount Band'].isnull().sum()}")
```

```python
print(df['Avg Sale per Unit'].head(10))
print(df['Total COGS'].head(10))
```

```python
# Limpar e converter Avg Sale per Unit
df['Avg Sale per Unit'] = (
    df['Avg Sale per Unit']
    .str.replace('€', '', regex=False)
    .str.replace(',', '.', regex=False)
    .str.strip()
    .astype(float)
)

# Limpar e converter Total COGS
df['Total COGS'] = (
    df['Total COGS']
    .str.replace('€', '', regex=False)
    .str.strip()
    .astype(float)
)

# Confirmar
print(df[['Avg Sale per Unit', 'Total COGS']].dtypes)
print(df[['Avg Sale per Unit', 'Total COGS']].head())
```

```python
df['Total COGS'] = (
    df['Total COGS']
    .str.replace('€', '', regex=False)
    .str.replace(',', '.', regex=False)
    .str.strip()
    .astype(float)
)

print(df[['Avg Sale per Unit', 'Total COGS']].dtypes)
print(df[['Avg Sale per Unit', 'Total COGS']].head())
```

```python
df.to_csv("financials_clean.csv", index=False)
print("Saved!")
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("financials_clean.csv")
df.head()
```

```python
df.shape()
```

```python
df.shape
```

```python
print(df['Segment'].head(10))
```

```python
subset = df[['Segment', 'Country', 'Sales', 'Profit']]
print(subset.head())
```

```python
government = df[df['Segment'] == 'Government']
print(f"Government rows: {len(government)}")
print(government[['Country', 'Sales', 'Profit']].head())
```

```python
gov_high = df[(df['Segment'] == 'Government') & (df['Sales'] > 1000000)]
print(f"Government + Sales > 1M: {len(gov_high)} rows")
```

```python
print(df.iloc[:3, :4])
```

```python
print(df.loc[10:12, ['Segment', 'Country', 'Sales']])
```

```python
top_sales = df.sort_values('Sales', ascending=False).head(10)
print(top_sales[['Segment', 'Country', 'Product', 'Sales', 'Profit']])
```

```python
print(df[['Sales', 'Profit', 'Profit Margin %']].head())
```

```python
df['Profit Margin %'] = round((df['Profit'] / df['Sales']) * 100, 2)
print(df[['Sales', 'Profit', 'Profit Margin %']].head())
```

```python
df['Profitable'] = np.where(df['Profit'] > 0, 'Yes', 'No')
print(df['Profitable'].value_counts())
```

```python
def classify_margin(margin):
    if margin > 20:
        return 'Excellent'
    elif margin > 10:
        return 'Good'
    elif margin > 0:
        return 'Low'
    else:
        return 'Loss'

df['Performance'] = df['Profit Margin %'].apply(classify_margin)
print(df['Performance'].value_counts())
```

```python
df['Revenue Category'] = pd.cut(
    df['Sales'],
    bins=[0, 50000, 100000, float('inf')],
    labels=['Low', 'Medium', 'High']
)
print(df['Revenue Category'].value_counts())
```

```python
df.to_csv("financials_clean.csv", index=False)
print("Saved!")
```
